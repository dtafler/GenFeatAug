```python
import torch 
import os
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
from utils.utils import samples_per_class


labels100 = torch.load('/mnt/dtafler/test/timm/vit_base_patch14_dinov2.lvd142m/cifar100_0.01/train/labels.pt')
spc100 = samples_per_class(labels100)
labels10 = torch.load('/mnt/dtafler/test/timm/vit_base_patch14_dinov2.lvd142m/cifar10_0.01/train/labels.pt')
spc10 = samples_per_class(labels10)


# Configure Matplotlib to use DejaVu Serif
plt.rc('font', family='serif')
plt.rc('font', serif='DejaVu Serif')
plt.rc('text', usetex=False)  # Disable LaTeX

# Define the figure width in points (1 * 15 cm in points)
fig_width_pt = 425.196
fig_width_in = fig_width_pt / 72.27
fig_height_in = fig_width_in / 2


# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(fig_width_in, fig_height_in))

# fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].bar(np.arange(10), spc10)
axs[0].set_title('CIFAR10 LT', fontsize=12)
axs[0].set_xlabel('Class Index', fontsize=10)
axs[0].set_ylabel('Samples')
axs[0].set_xticks(np.arange(10))
axs[0].set_xticklabels(np.arange(0, 10))


axs[1].bar(np.arange(100), spc100, width=0.5)
axs[1].set_title('CIFAR100 LT', fontsize=12)
axs[1].set_xlabel('Class Index', fontsize=10)
# axs[1].set_ylabel('Samples')

axs[0].tick_params(axis='both', which='major', labelsize=8)
axs[1].tick_params(axis='both', which='major', labelsize=8)

plt.tight_layout()
plt.savefig('LT.png', dpi=300)
```
